# Trabajo Final - Física

```python
import random
import math

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
        
def setPlot(a, b):
    width = a
    height = b
    plt.figure(figsize=(width, height))
```

# 1. Onda sinusoidal en una cuerda

## a) Parámetros de la función de onda

|               Variable                              |       Valores       |
|:----------------------------------------------------|:-------------------:|
| Número de onda (rad/m)                              |    [5,00 - 10,0]    | 
| Frecuencia angular (m/s)                            |    [2,00 - 8,00]    | 
| Ángulo de fase (rad)                                |    [0,00 - 3,14]    |
| Amplitud                                            |    [0,500 - 1,00]   |

```python
k = random.uniform(5.00, 10.00) # Número de onda
w = random.uniform(2.00,8.00) #  Frecuencia angular
rho = random.uniform(0.00,3.14) # Ángulo de fase
A = random.uniform(0.50,1.00) # amplitud

df = pd.DataFrame({
"Número de onda": [k],
"Frecuencia angular": [w],
"Ángulo de fase":[rho],
"Amplitud": [A]
})

df
```

## b) Gráfico Y vs X en t = 0,00s

```python
t = 0

x = np.linspace(0, 2, 10**5)
y = A * np.sin(k * x - w * t + rho)

setPlot(20, 5)
plt.title("Perfil de ondas Y vx X para t = 0.00s")
plt.ylabel('Y'); plt.xlabel('X')
plt.plot(x, y)
plt.show()
```

## c) Perfiles de onda 

```python
rhos = [0, np.pi/2, np.pi]

x = np.linspace(0, 2, 10**5)

setPlot(20, 5)
for r in rhos:
    plt.title("Perfiles de ondas superpuestas")
    plt.ylabel('Y'); plt.xlabel('X')
    y = A * np.sin(k * x - w * t + r)
    plt.plot(x, y)   
plt.show()
```

El ángulo de fase traslada las ondas de forma lateral.

# 2. Superposición de ondas


```python
rhos = [0, math.pi/4, math.pi/2, math.pi, math.pi*2]

x = np.linspace(0, 2, 10**5)

for r in rhos:
    setPlot(20, 5)
    y1 = A * np.sin(k * x - w * t + 0)
    y2 = A * np.sin(k * x - w * t + r)
    plt.title("Diferencia de fase: {:0.2f} radianes".format(r))
    plt.ylabel('Y')
    plt.xlabel('X')
    plt.plot(x, y1 + y2)  
    plt.plot(x, y1, c="red")
    plt.plot(x, y2, c="green")  
    plt.show()
```

Todas las ondas son constructivas, salvo aquella cuya diferencia en ángulo de fase es de pi radianes, la cual es destructiva.

# 3. Ondas estacionarias en una cuerda

```python
from matplotlib import animation, rc
from IPython.display import HTML

plt.rcParams['animation.writer'] = 'avconv'

fig, ax = plt.subplots()

ax.set_xlim(( 0, 1.5))
ax.set_ylim((-2, 2))

line, = ax.plot([], [], lw=2)
line2, = ax.plot([], [], lw=2)
line3, = ax.plot([], [], lw=2)

def init():
    line.set_data([], [])
    return (line,)

def animate(i):
    x = np.linspace(0, 1.5, 1000)
    y = np.sin( k * x - w * i)
    y2 = np.sin( k * x + w * i)
    line.set_data(x, y)
    line2.set_data(x, y2)
    line3.set_data(x, y + y2)
    return (line,)
```

La longitud de onda, para un armónico n, está dada por la siguiente ecuación:

$$\lambda_n = \frac{2(L)}{n}$$

Por lo tanto, los valores para lambda son:

```python
L = 1.5

n = [n for n in range(1, 5)]
lambs = [ 2 * L / n for n in n]

df = pd.DataFrame({
    "Armónico: ": n,
    "Longitud de onda": lambs
})

df
```

## Armónico n = 1

```python
l = 2 * L

k = 2 * np.pi / l
w = 2 * np.pi / 100

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=True)

HTML(anim.to_html5_video())
```
